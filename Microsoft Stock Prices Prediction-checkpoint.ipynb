# Importing libraries

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For calculation and web scrape
import math
import pandas_datareader as web

# For preprocessing and model building
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Style settings
sns.set()
plt.style.use('fivethirtyeight')
```

# Load and Explore the Dataset

```python
# Load the stock quote
df = web.DataReader('MSFT', data_source = 'yahoo', start = '2015-01-01', end = '2020-07-27')
```

```python
# Visualize the closing price history
plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD ($)',fontsize=18)
plt.show()
```

# Data Pre-processing

```python
# Create a new data frame to keep only the closing price using the 'Close' column
data = df.filter(['Close'])

# Converting the dataframe to a numpy array
dataset = data.values

# Create the training data set by getting the number of rows to train the model on. Recommended is at least 80% of original set.
training_data_len = math.ceil( len(dataset) *.8)

# Review data length
print(training_data_len)
```

```python
# Scale the all of the data to be values between 0 and 1 inclusive. Scaling is recommended when working with neural networks.
# Here, we used Data Normalization to accurately estimate min and max values - to stabilize sensitivity and weights of values.
scaler = MinMaxScaler(feature_range=(0, 1)) 
scaled_data = scaler.fit_transform(dataset)
```

```python
# Create training data set where the past 60 days closing prices values will be used to predict 61st closing price value.
# past 60 days: 'x_train’ 1st col with values of index 0 to 59 (60 values total) and 2nd col with values of index 1 to 60...
# target label: ‘y_train’ 1st col to contain the 61st value located at index 60 and the 62nd value located at index 61...

#Create the scaled training data set (all data from 0 to training_data_len)
train_data = scaled_data[0:training_data_len  , : ]

#Split the data to x_train and y_train
x_train=[]
y_train = []
for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
```

```python
# Convert x_train and y_train to numpy arrays for use in training the LSTM model
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
#Reshape the data into the shape accepted by the LSTM (3-dimensional - # of samples, time steps, features)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

# Build the Model

```python
#Build the LSTM network model
model = Sequential()

# First layer
model.add(LSTM(units=50, return_sequences=True,input_shape=(x_train.shape[1],1)))

# Second layer
model.add(LSTM(units=50, return_sequences=False))

# Third layer
model.add(Dense(units=25))

# Last layer
model.add(Dense(units=1))
```

```python
# Compile the model using the mean squared error (MSE) loss function and the adam optimizer
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

# Model Testing

```python
#Create a testing dataset
test_data = scaled_data[training_data_len - 60: , : ]

#Create the x_test and y_test data sets
x_test = []
y_test =  dataset[training_data_len : , : ] # 60 first actual prices, not scaled

for i in range(60,len(test_data)):
    x_test.append(test_data[i-60:i,0]) # Past 60 prices
```

```python
# Convert test data set (x_test) to a numpy array so we can use for testing the LSTM model
x_test = np.array(x_test)

# Check test data set shape
print(x_test)
```

```python
# Reshape the data into the shape accepted by the LSTM (3-dimensional - # of samples, time steps, features)
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
# Get the models predicted price values
predictions = model.predict(x_test)

# Get inverse shape of data to undo scaling
predictions = scaler.inverse_transform(predictions)
```

```python
# Calculate the value of Root Mean Squared Error (RMSE)
rmse=np.sqrt(np.mean(((predictions- y_test)**2)))

rmse
```

```python
# Plot/Create the data for the graph
train = data[:training_data_len]
actual_price = data[training_data_len:]
actual_price['Predictions'] = predictions

# Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(actual_price[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

# Another way to create the model

In this method, the training of the model is directly connected with the past 60 days closing price from the web.

```python
#Get the quote (actual prices)
microsoft_quote = web.DataReader('MSFT', data_source='yahoo', start='2015-01-01', end='2020-07-27')

#Create a new dataframe
new_df = microsoft_quote.filter(['Close'])

#Get the last 60 days closing price 
last_60_days = new_df[-60:].values

#Scale the data to be values between 0 and 1
last_60_days_scaled = scaler.transform(last_60_days)

#Create an empty list
X_test = []

# Append the past 60 days
X_test.append(last_60_days_scaled)

# Convert the X_test data set to a numpy array
X_test = np.array(X_test)

# Reshape the data
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Get the predicted scaled price
pred_price = model.predict(X_test)

# Undo the scaling 
pred_price = scaler.inverse_transform(pred_price)

# Show predicted price
print('Predicted price on 2020-07-27 is at', np.round(pred_price))
```
